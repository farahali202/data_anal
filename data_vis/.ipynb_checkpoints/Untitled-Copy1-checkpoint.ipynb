```python
y=data['Yearly Amount Spent']#output
x=data[['Time on App', 'Time on Website', 'Length of Membership']]#input
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
data.columns
```

# model building & training

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

# model prediction

```python

model.predict([[13,30,6]])#chhal mn ribh anrbh
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test.values
```

# model evaluation

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error#nisbat lkhataa
print(mean_absolute_error(y_test,y_pred))#wach 9rbt l zero wla la
```
